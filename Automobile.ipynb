9. Concatenate 2 Dataframes and create a key for the two Dataframes.

```python
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845,
171995, 135925 , 71400]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price':
[29995, 23600, 61500 , 58900]}
```

```python
GermanCarsDF = pd.DataFrame(GermanCars)
GermanCarsDF
```

```python
JapaneseCarsDF = pd.DataFrame(japaneseCars)
JapaneseCarsDF
```

```python
keys = ['German','Japanese']
CarsDF = pd.concat([GermanCarsDF,JapaneseCarsDF],keys = keys)
CarsDF
```

10. Create 2 Dataframes and merge them. Append the second Dataframe as a new column to the first Dataframe

```python
Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995,
135925 , 71400]}
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141,
80, 182 , 160]}
```

```python
Car_Price = pd.DataFrame(Car_Price)
Car_Price
```

```python
car_Horsepower = pd.DataFrame(car_Horsepower)
car_Horsepower
```

```python
result_df = pd.merge(Car_Price,car_Horsepower,on = 'Company')
result_df
```
